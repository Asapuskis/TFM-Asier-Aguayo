Carga del fichero json con los username de Discogs y elimininaci√≥n de duplicados para buscar sus ratings.

```python
import pandas as pd
```

```python
usernameRatingsCSV = '../TFM Files/usernameRatingsMerged.csv'

dfRatings = pd.read_csv(usernameRatingsCSV)
dfRatings
```

```python
print("N√∫mero de usuarios (con repetidos): ", dfRatings.count())
dfUsersRatings = dfRatings.drop_duplicates("user_id")
print("N√∫mero de usuarios: ", dfUsersRatings.count())
```

```python
print("N√∫mero de items (con repetidos): ", dfRatings.count())
dfItemsRatings = dfRatings.drop_duplicates("master_id")
print("N√∫mero de items: ", dfItemsRatings.count())
```

<p>La cantidad de diferentes items es demasiado grande para calcularla con un equipo no preparado para calcular grandes dimensiones.</p>
<p>Por eso, se van a utilizar los ratings que se pueden encontrar en los datos utilizados por el filtrado de contenido.</p>

```python
fullJsonFile = '../TFM Files/CNT2.2 - dataScaled/dataScaled.json'

scaledData = pd.read_json(path_or_buf=fullJsonFile, lines=True)
    
scaledData
```

```python
master_idList = scaledData["master_id"].tolist()
master_idList
```

```python
dfFiltered = dfRatings[dfRatings.master_id.isin(master_idList)]
dfFiltered
```

<p>Mostramos la cantidad de votos que tienen las publicaciones</p>

```python
dfCounts = dfFiltered.groupby(["master_id"]).agg("size").to_frame('count').reset_index()
dfCounts
```

<p>Ordenamos para ver cuales son las 20 publicaciones m√°s veces valoradas.</p>

```python
dfCounts.sort_values("count", ascending=False).head(20)
```

<p>Calculamos la media de las votaciones</p>

```python
dfMean = dfFiltered.groupby(["master_id"]).agg("mean")
dfMean
```

<p>Creamos el archivo con ratings filtrado.</p>

```python
dfFiltered.to_csv(path_or_buf = "../TFM Files/usernameRatingsFiltered.csv" , index=False)
```

<p>Tambi√©n creamos un csv con la relaci√≥n entre id y nombre de las publicaciones.</p>

<p>Primero ponemos en com√∫n los ids de los que tenemos ratings.</p>

```python
master_idRatedList = dfFiltered.drop_duplicates("master_id")

master_idRatedList = master_idRatedList["master_id"].tolist()
master_idRatedList
```

```python
dfFilteredScaledData = scaledData[scaledData.master_id.isin(master_idRatedList)]
dfFilteredScaledData
```

<p>Finalmente, guardamos: </p>

1. una tabla con la relaci√≥n entre ids y nombre de la publicaci√≥n.
2. La tabla con los datos filtrada con los ids de las publicaciones de las que disponemos ratings.

```python
dfFilteredScaledData[['master_id','title']].to_csv(path_or_buf = "../TFM Files/idTitleList.csv" , index=False)
```

```python
dfFilteredScaledData.to_json(path_or_buf = "../TFM Files/dataScaledFiltered.json", orient='records', lines=True)
```
